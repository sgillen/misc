```python
import scipy.signal as sgn
import numpy as np
from matplotlib.pyplot import plot, figure
from scipy.fftpack import fft, fftfreq

%matplotlib inline
```

```python
t = np.arange(-10,10,.1)
#x = np.random.normal(size = t.shape[-1])
#x = np.sin(t)
#x = np.cos(t)
#x = np.sinc(t)

x = np.ones(t.size) + np.sin(t)
y = sgn.hilbert(x)
a = x + 1j*y
```

```python
#figure(0)
plot(x)
figure()
#plot(np.real(y))
#figure(2)
#plot(np.imag(y))
#figure()

plot(np.real(a))
figure()
plot(np.imag(a))
figure()
plot(np.abs(a))
```

So we can see (or not see because I'm lazy with my plots) that real part of the time domain hilbert transform is our original signal, and the complex part is doing something weird

Now lets look the frequency domain

```python
#X = np.fft.fft(x)
#Y = np.fft.fft(y)

X = fft(x)
Y = fft(y)
A = fft(a)
freq = fftfreq(t.shape[-1])
```

```python
figure()
plot(freq, abs(X))
figure()
plot(freq, abs(Y))
figure()
plot(freq, abs(A))

```
