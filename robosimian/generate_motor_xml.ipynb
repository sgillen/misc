This script just generates the urdf for motors that drake will understand. I took this and copy pasted the output into robosim_nomesh.urdf. requires the lxml library which is not actually part of the standard library but seemed to come packaged with the default ubuntu install of python anyway. 

```python
import lxml.etree as ET
#import xml.etree.cElementTree as ET  #this is the built in parser, it works fine too but doesn't do any pretty print anything

root = ET.Element("root")

num_joints = 8
num_limbs = 4

for limb_index in range(num_limbs):
    limb_str = "limb" + str(limb_index)
    for joint_index in range(num_joints):
        joint_str = "joint" + str(joint_index)
        
        joint_name = limb_str + "_" + joint_str
        
        transmission = ET.SubElement(root, "transmission", type="SimpleTransmission", name=(joint_name + "_" + "trans"))
        ET.SubElement(transmission, "actuator", name=joint_name)
        ET.SubElement(transmission, "joint", name=joint_name)
        ET.SubElement(transmission, "mechanicalReduction").text = "1"
        
        
        
ET.dump(root)
```
