```python
disp = np.array([0, .4 , .8 , 1.2, 1.6])
force = np.array([0, .04, .16, .4, 1.8])
```

```python
plt.plot(disp, force)
```

```python
def model_func(x,a,b,c):
    return a*np.exp(b*x) + c
```

```python
popt, pcov = curve_fit(model_func, disp, force)
print(popt)
```

```python
x_data = np.linspace(disp[0], disp[-1], 100)
plt.plot(x_data, model_func(x_data, *popt))
plt.plot(disp, force, 'x-')
```

```python
fsolve(lambda x: model_func(x, *popt) - 10e-6 , .1, )
```
