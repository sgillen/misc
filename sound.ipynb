```python
import sounddevice as sd
from numpy import *
from matplotlib.pyplot import *

%matplotlib inline
```

```python
fs = 44100       # sampling rate, Hz
duration = 60.0   # in seconds
f = 75      # sine frequency, Hz

y = abs((sin(2*pi*arange(fs*duration)*f/fs)).astype(np.float32))
```

```python
#plot(y[1:1000])
```

```python
sd.play(y,fs)
sd.stop()
```
