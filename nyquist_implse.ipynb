```python
import numpy as np
import scipy as sp
from scipy.fftpack import fft
from matplotlib.pyplot import plot, figure, stem

%matplotlib inline

N = .1
t = np.arange(1,10,N)
x = np.sin(t)

i_train = np.ones(t.size)
i_train[0::2] = -1

stem(i_train[1:10])
figure()
plot(x)
figure()

x_mul = i_train*x
plot(x_mul)
figure()
x_f = fft(x)
i_f = fft(i_train)
xi_f = fft(x_mul)

#axes = plt.gca()
#axes.set_xlim([xmin,xmax])
#axes.set_ylim([ymin,ymax])

freq = sp.fftpack.fftfreq(t.shape[-1], N )

plot(freq, np.abs(x_f), 'o')
figure()
plot(freq, np.abs(i_f), 'o')
figure()
plot(freq, np.abs(xi_f), 'o')

```
